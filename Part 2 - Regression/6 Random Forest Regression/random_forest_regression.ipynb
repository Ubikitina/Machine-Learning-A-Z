# Random Forest Regression

Random Forest is a type of ensemble learning, where multiple decision trees are combined to make a stronger prediction. Here's how it works:

1. A subset of data points is randomly selected from the dataset, and a decision tree is built for that subset.
2. This process is repeated to build many decision trees (often 500 or more).
3. For a new data point, each tree makes a prediction, and the final prediction is the average of all tree predictions.

image.png

Benefits:
- Random Forest **improves accuracy** by averaging the predictions of many trees, reducing the impact of errors or biases in individual trees.
- The ensemble method also makes it **more stable**, as changes in data are less likely to significantly affect the overall result.

Analogy of guessing the number of jellybeans in a jar: where averaging guesses from many people leads to a more accurate result. Similarly, Random Forest uses multiple decision trees to enhance prediction accuracy.

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
from google.colab import drive
drive.mount('/content/drive')
```

```
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning A-Z/Part 2 - Regression/6 Random Forest Regression/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values # matrix of features
y = dataset.iloc[:, -1].values # dependent variable vector
```

## Training the Random Forest Regression model on the whole dataset

```
# Import the RandomForestRegressor class from sklearn
from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest regressor object with 10 trees (n_estimators) and set random_state for reproducibility
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)

# Fit the regressor to the training data (X: features, y: target variable)
regressor.fit(X, y)
```

## Predicting a new result

```
# Predict the target value for a new data point where the feature value is 6.5
regressor.predict([[6.5]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
# Create a grid of values from the minimum to the maximum of X, with a step size of 0.01 for smooth curve plotting
X_grid = np.arange(min(X), max(X), 0.01)

# Reshape X_grid into a 2D array where each value is in its own row (necessary for prediction and plotting)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
