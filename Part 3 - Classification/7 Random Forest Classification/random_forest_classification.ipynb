# Random Forest Classification

Random Forest Classification is an ensemble learning technique that combines multiple decision trees to improve the overall predictive performance of the model. Instead of relying on a single decision tree, the algorithm constructs multiple trees, each trained on a random subset of the data. This approach helps reduce errors and improves accuracy by leveraging the "wisdom of the crowd."

**Ensemble Learning**:

This concept involves combining multiple machine learning algorithms to create a more robust final model. In the case of Random Forest, it uses many decision trees to make a classification.


**Random Forest Classification Step-by-Step**:

Hereâ€™s how Random Forest Classification works:

1. **Random Data Subsets**: At the start, the algorithm selects random subsets of data from the training set. This ensures that each decision tree is trained on slightly different data, introducing variability in the predictions.

2. **Building Decision Trees**: For each subset, a decision tree is built. Instead of using the entire dataset, each tree is only trained on a portion of the data, making them unique but slightly less accurate when used alone.

3. **Predicting with Multiple Trees**: When classifying a new data point, each of the decision trees provides its own prediction. The final classification is determined by the majority vote across all the trees. The class that most trees predict becomes the final prediction.



image.png

**Improved Accuracy**:

Even if individual decision trees are imperfect, the ensemble of trees improves overall performance by averaging out the noise and errors from any single tree. This process reduces overfitting and enhances accuracy.

**Real-world example**:

A real-world example of Random Forest's application is in Microsoft's Kinect system, where it was used to track human body movements in real-time. The algorithm helped process data faster and more efficiently, leading to lower hardware costs.

Overall, Random Forest Classification is highly effective for tasks that require precise and robust predictions, such as image recognition or, in the case of Kinect, human motion detection.

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning A-Z/Part 3 - Classification/7 Random Forest Classification/Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values # matrix of features
y = dataset.iloc[:, -1].values # dependent variable vector
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split

# Split data into 75% training and 25% testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
print(y_train)
```

```
print(X_test)
```

```
print(y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# Fit the scaler to X_train and transform X_train to standardize it (mean=0, std=1)
X_train = sc.fit_transform(X_train)

# Transform X_test using the same scaler (fitted on X_train) to ensure consistent scaling
X_test = sc.transform(X_test)
```

```
print(X_train)
```

```
print(X_test)
```

## Training the Random Forest Classification model on the Training set

```
# Imports the RandomForestClassifier class from sklearn's ensemble module, which implements the Random Forest algorithm
from sklearn.ensemble import RandomForestClassifier

# Creates an instance of the RandomForestClassifier:
# n_estimators=10: Specifies the number of trees in the forest.
# criterion='entropy': Uses entropy as the criterion for measuring the quality of a split (alternative is 'gini').
# random_state=0: Ensures reproducibility by setting a seed for the random number generator.
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)

# Trains the Random Forest classifier on the training data (X_train and y_train).
classifier.fit(X_train, y_train)
```

## Predicting a new result

```
# Predict the class (0 or 1) for a new data point with Age=30 and EstimatedSalary=87000, after scaling it
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```
# Predict the target labels for the test set using the trained classifier
y_pred = classifier.predict(X_test)

# Concatenate the predicted labels (y_pred) and the actual test labels (y_test) into a single array for comparison
# Reshape both y_pred and y_test into column vectors and concatenate them horizontally (axis=1)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score

# Compute the confusion matrix to compare predicted and actual labels in the test set
cm = confusion_matrix(y_test, y_pred)

# Print the confusion matrix, which shows true positives, true negatives, false positives, and false negatives
print(cm)

# Calculate and return the accuracy score, which is the proportion of correctly predicted labels
accuracy_score(y_test, y_pred)
```

The confusion matrix:

- **63**: True negatives (correctly predicted class 0)
- **28**: True positives (correctly predicted class 1)
- **5**: False positives (incorrectly predicted as class 1, but actually class 0)
- **4**: False negatives (incorrectly predicted as class 0, but actually class 1)

The **accuracy score** of **0.91** means that the model correctly predicted 91% of the test set, reflecting a high level of accuracy in this classification task.

We are surprised as we initially believed we had a chance to surpass a previous accuracy record. We did not break the record of 93%, achieving 91% accuracy instead.

However, we encourage experimenting with the classification models by adjusting parameters to potentially beat the 93% accuracy.

## Visualising the Training set results

```
from matplotlib.colors import ListedColormap

# Reverse the scaling on the training data to get the original feature values (Age, Estimated Salary)
X_set, y_set = sc.inverse_transform(X_train), y_train

# Create a mesh grid for plotting the decision boundary, defining the range for Age and Estimated Salary
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

# Plot the decision boundary by classifying each point on the grid using the trained classifier
# Use `contourf` to fill the background regions with colors corresponding to each class (0 or 1)
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))

# Set the x and y axis limits to the min and max values of the mesh grid
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plot the training data points with different colors for each class
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('Random Forest Classification (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_test), y_test

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('Random Forest Classification (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
