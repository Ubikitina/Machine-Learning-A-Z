# Support Vector Machine (SVM)

Support Vector Machines (SVM) are a type of supervised machine learning algorithm primarily used for classification tasks. The core idea of SVM is to find a decision boundary (called a hyperplane in higher dimensions) that best separates different classes of data points.

### Key Concepts:

1. **Separating the Data**:
   - In a two-dimensional space, you have points representing two different categories (e.g., red and green points).
   - The goal is to draw a line (or in higher dimensions, a hyperplane) that divides these points into two distinct classes, so that when new data is introduced, the model can predict which class the new point belongs to.

2. **Multiple Decision Boundaries**:
   - Several possible lines can separate the data, but each would classify future points differently. The challenge is to find the "optimal" line that provides the best separation.

3. **Maximum Margin**:
   - SVMs select the line that has the **maximum margin**, meaning the line that is as far as possible from the nearest points of each class. This margin ensures that future predictions are more robust.
   - The margin is defined as the distance between the line (hyperplane) and the nearest data points from each class.

4. **Support Vectors**:
   - The points that are closest to the decision boundary (the hyperplane) are called **support vectors**.
   - These are the most critical data points for the SVM, as they determine the position of the decision boundary. All other points donâ€™t influence the result of the SVM.

5. **Hyperplanes in Higher Dimensions**:
   - In more complex datasets with many features (e.g., 3, 5, 10 dimensions), the decision boundary becomes a hyperplane, a generalized version of a line. It separates the data points in these higher-dimensional spaces.

6. **Why SVM is Special**:
   - Instead of focusing on the most typical (core) examples of each class, SVM looks at the **extreme cases**â€”the points closest to the boundary (the support vectors), which are more challenging to classify.
   - This makes SVM a more "risky" or "rebellious" algorithm, but it often leads to better performance, especially with complex datasets.

In essence, SVM works by maximizing the margin between the two classes, relying on the support vectors (the critical points near the boundary), and providing a robust way of classifying new, unseen data based on this optimal separation.

The **Social_Network_Ads.csv** dataset contains three columns:

1. **Age**: The age of the individual (e.g., 19, 35, 26).
2. **EstimatedSalary**: The estimated salary of the individual (e.g., 19,000, 20,000, 43,000).
3. **Purchased**: A binary outcome (0 or 1), indicating whether the individual made a purchase (0 = did not purchase, 1 = purchased).

This dataset is commonly used for classification tasks, such as predicting whether a person will make a purchase based on their age and estimated salary.

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split

# Split data into 75% training and 25% testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
print(y_train)
```

```
print(X_test)
```

```
print(y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# Fit the scaler to X_train and transform X_train to standardize it (mean=0, std=1)
X_train = sc.fit_transform(X_train)

# Transform X_test using the same scaler (fitted on X_train) to ensure consistent scaling
X_test = sc.transform(X_test)
```

```
print(X_train)
```

```
print(X_test)
```

## Training the SVM model on the Training set

```
# Import the Support Vector Classifier (SVC) from scikit-learn
from sklearn.svm import SVC

# Create an instance of the SVM classifier with a linear kernel
# 'linear' specifies that we want a linear decision boundary (hyperplane)
# 'random_state = 0' ensures reproducibility of the results by fixing the seed
classifier = SVC(kernel = 'linear', random_state = 0)

# Fit the classifier to the training data
# X_train: the training features
# y_train: the corresponding training labels
classifier.fit(X_train, y_train)
```

## Predicting a new result

```
# Predict the class (0 or 1) for a new data point with Age=30 and EstimatedSalary=87000, after scaling it
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```
# Predict the target labels for the test set using the trained classifier
y_pred = classifier.predict(X_test)

# Concatenate the predicted labels (y_pred) and the actual test labels (y_test) into a single array for comparison
# Reshape both y_pred and y_test into column vectors and concatenate them horizontally (axis=1)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score

# Compute the confusion matrix to compare predicted and actual labels in the test set
cm = confusion_matrix(y_test, y_pred)

# Print the confusion matrix, which shows true positives, true negatives, false positives, and false negatives
print(cm)

# Calculate and return the accuracy score, which is the proportion of correctly predicted labels
accuracy_score(y_test, y_pred)
```

The confusion matrix:

- **66**: True negatives (correctly predicted class 0)
- **24**: True positives (correctly predicted class 1)
- **2**: False positives (incorrectly predicted as class 1, but actually class 0)
- **8**: False negatives (incorrectly predicted as class 0, but actually class 1)

The **accuracy score** of **0.9** means that the model correctly predicted 93% of the test set, reflecting a high level of accuracy in this classification task.

The SVM model did not outperform the k-Nearest Neighbors (kNN) model because a **linear kernel** was used. This means the decision boundary was a straight line, which limited its ability to accurately classify certain points. Rotating the linear boundary still resulted in incorrect predictions, as it couldn't properly separate the green points (representing one class) from the red points (another class). A non-linear boundary, such as a curved one, would be more effective in this scenario, as it would better separate the red and green regions.

## Visualising the Training set results

```
from matplotlib.colors import ListedColormap

# Reverse the scaling on the training data to get the original feature values (Age, Estimated Salary)
X_set, y_set = sc.inverse_transform(X_train), y_train

# Create a mesh grid for plotting the decision boundary, defining the range for Age and Estimated Salary
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

# Plot the decision boundary by classifying each point on the grid using the trained classifier
# Use `contourf` to fill the background regions with colors corresponding to each class (0 or 1)
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))

# Set the x and y axis limits to the min and max values of the mesh grid
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plot the training data points with different colors for each class
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_test), y_test

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
