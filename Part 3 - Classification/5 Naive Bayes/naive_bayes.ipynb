# Naive Bayes

Bayes' Theorem is a fundamental concept in probability that helps us calculate the probability of an event given prior knowledge of related events. To understand it better, let’s walk through the given scenario involving two machines that produce wrenches (also called spanners outside North America) in a factory.

### The Setup:
- **Machine 1** produces 30 wrenches per hour, and **Machine 2** produces 20 wrenches per hour. So, in total, 50 wrenches are produced per hour.
- **1% of all wrenches are defective**.
- **50% of the defective wrenches** come from Machine 1, and **50% from Machine 2**.

### The Question:
What is the probability that a wrench produced by Machine 2 is defective?

To answer this, we’ll use Bayes’ Theorem, which helps us reverse the probability relationship.

### Bayes' Theorem Formula:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

In this case:
- **A** is the event that the wrench is defective.
- **B** is the event that the wrench came from Machine 2.

Thus, the formula becomes:

$$
P(\text{Defective | Machine 2}) = \frac{P(\text{Machine 2 | Defective}) \cdot P(\text{Defective})}{P(\text{Machine 2})}
$$

This would read: the probability that the wrench is defective given that it has been produced by machine 2.

Let’s break this down step-by-step:

- Step 1: **Probability that a randomly selected wrench is from Machine 2 (P(Machine 2)):** Machine 2 produces 20 wrenches per hour, and 50 wrenches are produced in total. So, the probability of picking a wrench from Machine 2 is:

$$
P(\text{Machine 2}) = \frac{20}{50} = 0.4
$$

- Step 2: **Probability that a wrench is defective (P(Defective))**: We are given that 1% of all wrenches are defective, so:

$$
P(\text{Defective}) = 0.01
$$

- Step 3: **Probability that a defective wrench came from Machine 2 (P(Machine 2 | Defective))**: Out of all defective wrenches, 50% come from Machine 2, so:

$$
P(\text{Machine 2 | Defective}) = 0.5
$$

- Step 4: **Applying Bayes’ Theorem and getting the Result**: Now, let’s plug these values into the formula:

$$
P(\text{Defective | Machine 2}) = \frac{0.5 \cdot 0.01}{0.4} = \frac{0.005}{0.4} = 0.0125
$$

Thus, the probability that a wrench from Machine 2 is defective is **0.0125**, or **1.25%**.

### Intuitive Explanation:
Machine 2 produces fewer wrenches (40% of total production), but it contributes to 50% of the defective wrenches. This slight discrepancy shows that wrenches from Machine 2 are more likely to be defective compared to those from Machine 1. Bayes’ Theorem helps us combine this information logically, allowing us to calculate the probability of a defect given that the wrench came from Machine 2.

This simple example illustrates how Bayes' Theorem works in practical scenarios where we have prior information and want to compute conditional probabilities based on that information.

### Walk and Drive Example

We will apply Bayes' Theorem to classify data points based on their features.

**Dataset and Features**:
- The dataset includes two features: age and salary.
- Two categories are defined: "walks to work" (red) and "drives to work" (green).




image.png

**Classification Process**:

**Step 1: Calculate Probabilities for Walking**:
- **Prior Probability**: The probability that a person walks to work is calculated by dividing the number of people who walk by the total number of people.
  - Example: There are 10 people who walk and 30 total people. Thus,

$$
P(\text{Walk}) = \frac{10}{30} \approx 0.33 \text{ or } 33\%
$$


- **Marginal Likelihood**: This is the probability of a new observation having similar features. A radius is chosen around the new data point to include similar data points.
  - For instance, with a radius defined around age 25 and salary $30,000, the marginal likelihood might be calculated as the number of similar observations divided by the total number of observations.
  - Example: If 4 out of 30 total observations fall within this radius,

$$
P(X) = \frac{4}{30} \approx 0.133 \text{ or } 13.3\%
$$


- **Likelihood**: probability that a randomly selected data point with the given features belongs to the "walks" category.
  - Example: Among the people who walk, if 3 out of 10 have features similar to the new data point,

$$
P(X|\text{Walk}) = \frac{3}{10} = 0.3 \text{ or } 30\%
$$


- **Posterior Probability**: probability that a new observation belongs to the "walks" category given its features is calculated using Bayes' Theorem:

$$
P(\text{Walk}|X) = \frac{P(X|\text{Walk}) \times P(\text{Walk})}{P(X)}
$$


- With the values provided:

$$
P(\text{Walk}|X) = \frac{0.3 \times 0.33}{0.133} = 0.75 \text{ or } 75\%
$$



image.png

**Step 2: Calculate Probabilities for Driving**:

Similar calculations are done for the "drives to work" category:
- **Prior Probability**: If there are 20 people who drive,

$$
P(\text{Drive}) = \frac{20}{30} = 0.67 \text{ or } 67\%
$$


- **Marginal Likelihood**: Suppose 6 out of 30 observations fall within the radius,

$$
P(X) = \frac{4}{30} = 0.13 \text{ or } 13\%
$$


- **Likelihood**: If 1 out of 20 people who drive has similar features,

$$
P(X|\text{Drive}) = \frac{1}{20} = 0.05 \text{ or } 5\%
$$


- **Posterior Probability**:

$$
P(\text{Drive}|X) = \frac{0.05 \times 0.67}{0.13} = 0.25 \text{ or } 25\%
$$



image.png

**Step 3: Comparison and Classification**:
  - The posterior probability for walking is 75%, and for driving is 25%.
  - Since 75% is greater than 25%, the new data point is classified as "walks to work."


### Some Extra Considerations

- **Why it's called "Naive" Bayes**:
  - The algorithm assumes that features are independent, which is often not true in reality, but still produces good results.
  - Example: age and salary, in practice, these variables may be correlated.

- **"P(X)" (the likelihood of a new data point having certain features)**:
  - The same P(X) value is used when comparing two different outcomes, like walking to work or driving. In both cases, this P(X) is used at the bottom of the division (denominator).
  - When comparing outcomes, the denominator in the probability formula can be dropped, simplifying calculations.
  - Note: we can only remove it if our objective is to compare the values. If we want to calculate the final probability, then we should not remove the divisor.



image.png

- **When there are more than two classes in a dataset:** It follows a similar/straight process as it always adds up to 1 (e.g. with 2 classes: 0.75 + 0.25 = 1). If we have 3 classes and classify 1 and it's greater than 50% we can assign that class, but if not we have to calculate for each to assign the classification.

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
from google.colab import drive
drive.mount('/content/drive')
```

```
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning A-Z/Part 3 - Classification/5 Naive Bayes/Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values # matrix of features
y = dataset.iloc[:, -1].values # dependent variable vector
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split

# Split data into 75% training and 25% testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
print(X_train)
```

```
print(y_train)
```

```
print(X_test)
```

```
print(y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# Fit the scaler to X_train and transform X_train to standardize it (mean=0, std=1)
X_train = sc.fit_transform(X_train)

# Transform X_test using the same scaler (fitted on X_train) to ensure consistent scaling
X_test = sc.transform(X_test)
```

```
print(X_train)
```

```
print(X_test)
```

## Training the Naive Bayes model on the Training set

```
# Import the Gaussian Naive Bayes classifier from sklearn
from sklearn.naive_bayes import GaussianNB

# Create an instance of the Gaussian Naive Bayes classifier
classifier = GaussianNB()

# Train the classifier using the training data (X_train: features, y_train: labels)
classifier.fit(X_train, y_train)
```

## Predicting a new result

```
# Predict the class (0 or 1) for a new data point with Age=30 and EstimatedSalary=87000, after scaling it
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```
# Predict the target labels for the test set using the trained classifier
y_pred = classifier.predict(X_test)

# Concatenate the predicted labels (y_pred) and the actual test labels (y_test) into a single array for comparison
# Reshape both y_pred and y_test into column vectors and concatenate them horizontally (axis=1)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score

# Compute the confusion matrix to compare predicted and actual labels in the test set
cm = confusion_matrix(y_test, y_pred)

# Print the confusion matrix, which shows true positives, true negatives, false positives, and false negatives
print(cm)

# Calculate and return the accuracy score, which is the proportion of correctly predicted labels
accuracy_score(y_test, y_pred)
```

The confusion matrix:

- **65**: True negatives (correctly predicted class 0)
- **25**: True positives (correctly predicted class 1)
- **3**: False positives (incorrectly predicted as class 1, but actually class 0)
- **7**: False negatives (incorrectly predicted as class 0, but actually class 1)

The **accuracy score** of **0.9** means that the model correctly predicted 90% of the test set, reflecting a high level of accuracy in this classification task.

The accuracy we get with that naive bayes model is 90% which beats indeed logistic aggression but does equally the same as the classic SVM model with a linear kernel.

## Visualising the Training set results

```
from matplotlib.colors import ListedColormap

# Reverse the scaling on the training data to get the original feature values (Age, Estimated Salary)
X_set, y_set = sc.inverse_transform(X_train), y_train

# Create a mesh grid for plotting the decision boundary, defining the range for Age and Estimated Salary
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

# Plot the decision boundary by classifying each point on the grid using the trained classifier
# Use `contourf` to fill the background regions with colors corresponding to each class (0 or 1)
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))

# Set the x and y axis limits to the min and max values of the mesh grid
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plot the training data points with different colors for each class
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('Naive Bayes (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_test), y_test

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('Naive Bayes (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
