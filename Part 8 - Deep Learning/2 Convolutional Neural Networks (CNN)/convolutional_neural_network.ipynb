# Convolutional Neural Network

## What are Convolutional Neural Netorks?

Optical illusions demonstrate how the human brain identifies and categorizes features in images, emphasizing the importance of feature detection in perception. **Convolutional Neural Networks (ConvNets)** mimic this feature-based classification, showcasing examples where neural networks accurately or sometimes incorrectly classify images, much like humans do. The practical applications of ConvNets is image recognition in self-driving cars and social media platforms like Facebook (CNN have now gained a lot of popularity, even over the ANNs.).

Geoffrey Hinton and Yann LeCun are key figures in the field. They contributed to the development and advancement of deep learning and ConvNets.

ConvNets operate by using a digital representation of images as arrays of pixel values. The process involves several steps:
1. **Convolution**: This process involves applying a filter (kernel) to an image to detect specific features like edges, textures, or patterns.
2. **Max Pooling**: This operation reduces the spatial dimensions of an image by taking the maximum value from smaller regions, helping to retain important features while reducing computation.
3. **Flattening**: This step converts the pooled feature maps into a single-dimensional vector, preparing the data for input into a fully connected neural network layer.
4. **Full Connection**: In this stage, the flattened vector is passed through fully connected layers where neurons are connected to every activation in the previous layer, enabling the network to make final predictions.

These steps enable the network to extract and interpret features from images to make accurate classifications. Example:

### Step 1 - Convolution Operation and ReLU Layer

#### Convolution Operation

Convolution is the process of integrating two functions to see how one modifies the other. This is a familiar concept in fields like signal processing, where the equation shown below is very popular:

Convolution can be described intuitively by using a **Feature Detector** (also known as a **Kernel** or **Filter**) applied to an input image.
- Typically, a 3x3 feature detector is used to scan the image with a stride of 1 pixel, though larger strides like 2 are common for reducing image size and computational load.
- The feature detector performs element-wise multiplication with the image pixels it covers, and the results are summed to produce a feature map, which highlights specific features such as edges.

Multiple feature maps are generated using different filters, allowing the network to detect various features simultaneously. Examples of different filters (e.g., Sharpen, Blur, Edge Detect) demonstrate how feature maps can emphasize different aspects of an image. Although some filters like Emboss may not be directly relevant to ConvNets, others like Edge Detect are crucial for identifying important features.

The key takeaway is that convolution helps in extracting and preserving essential features from images while reducing their size, facilitating efficient processing by the neural network. This feature extraction mirrors human visual perception, where the brain focuses on significant features rather than every individual pixel.

#### ReLU Layer

ReLU or Rectified Linear Unit is a simple yet powerful activation function that plays a critical role in enhancing the non-linear processing abilities of convolutional neural networks. By selectively filtering out negative values and preserving essential features, allows the network to capture more intricate features within the data.

The integration of ReLU enhances the network's ability to recognize and differentiate between various elements in an image, such as edges, shadows, and textures. For instance, when a feature detector identifies a pattern that transitions smoothly from light to dark, ReLU disrupts this linear progression by eliminating negative values, thereby emphasizing abrupt changes and preserving significant features. This process mirrors the human visual system, which focuses on important aspects of an image rather than processing every single pixel.

Multiple feature maps are generated using different filters, each highlighting specific characteristics of the input image. ReLU plays a pivotal role in ensuring that these feature maps retain their spatial relationships while reducing the complexity of the data. By doing so, it not only facilitates efficient processing but also enhances the network's performance in tasks such as image classification and object detection.

### Step 2 - Pooling

**Max pooling** is aimed at enhancing the network's ability to recognize objects regardless of their orientation and position within an image. By applying max pooling, the neural network achieves spatial invariance, enabling it to identify features like a cheetah's distinctive patterns even when the cheetah appears rotated, squashed, or positioned differently across various images.

The process of max pooling involves **selecting the maximum value from a specific region of the feature map**, typically using a 2x2 pixel window with a stride of two. This operation effectively reduces the spatial dimensions of the image by **retaining only the most significant features and discarding the rest**.
> *For example, in a 2x2 window, out of four pixel values, only the highest value is preserved, which helps in minimizing the amount of data the network needs to process while maintaining critical information.*

Max pooling offers several key benefits:
1. **Spatial Invariance**: It allows the network to recognize features irrespective of their exact location within the image.
2. **Dimensionality Reduction**: By decreasing the size of the feature maps, max pooling reduces the computational load and the number of parameters in the network.
3. **Prevention of Overfitting**: By eliminating less important information, max pooling helps in preventing the model from becoming too tailored to the training data, thus enhancing its generalization capabilities.

There are different types of pooling, such as mean pooling and sum pooling, but max pooling is the most effective in preserving essential features. Practical examples demonstrate how max pooling maintains important attributes of the input image while significantly reducing its size.

### Step 4 - Full Connection

In this step, we add a whole artificial neural network (ANN) to to the CNN to optimize predictions further.

This is how the ANN looks:
- **Input Layer**: joins the stage resulting from flattening.
- **Fully connected layer**: hidden layers that connect all neurons. In CNNs, the hidden layer must be fully connected.
- **Output Layer**: same as in ANN. For example, a network classifying between a dog and a cat would have two output neurons, each associated with one class.


image.png

When training, the network learns **which features** (e.g., "dog features" like floppy ears) **are most significant** by adjusting weights and feature detectors. The **error** is calculated and minimized using a loss function, and adjustments are made via **backpropagation**. As the network trains, it optimizes the feature detectors to focus on the most relevant attributes, iterating over many epochs.

Ultimately, once trained, the network can classify new images based on learned features, **calculating probabilities for each class** based on the presence of these key features. This process allows the network to **predict classes with high accuracy** by leveraging optimized feature detection and weight adjustments.

### Softmax & Cross-Entropy

The softmax function is essential for transforming CNN outputs into a valid probability distribution for classification tasks. Paired with the cross-entropy loss function, it enhances the training process by providing effective gradients for optimization.

**Softmax Function**

- **Purpose:** Converts the raw output scores (logits) from the final layer of a CNN into probabilities that sum to 1. For example, it might output 95% for a dog and 5% for a cat.
- **Mechanism:** Applies the exponential function to each output score and then normalizes these values by dividing by the sum of all exponentials. This ensures all probabilities are between 0 and 1 and collectively add up to 1.
- **Importance:** Without softmax, the output neurons operate independently and their probabilities wouldn't necessarily sum to 1, making the predictions incoherent for classification tasks.

**Cross-Entropy Loss Function**
- **Purpose:** Measures the difference between the predicted probability distribution and the actual distribution (labels). It serves as a loss function to be minimized during training.
- **Comparison with Mean Squared Error (MSE):**
  - **Classification vs. Regression:** Cross-entropy is better suited for classification problems, especially when combined with softmax, whereas MSE is typically used for regression tasks.
  - **Gradient Behavior:** Cross-entropy provides stronger gradients when predictions are far from actual values, facilitating more effective weight updates during backpropagation. MSE can result in very small gradients when predictions are already close, slowing down learning.
- **Example:** When comparing two neural networks, cross-entropy can better differentiate their performance even if both have the same classification error rate. It provides a more nuanced measure of how confident the networks are in their predictions.

## Part 1 - Data Preprocessing

### Preprocessing the Training set

The below code prepares us to do **Image Augmentation**, which involves applying various transformations to images in a training set to increase their diversity. These transformations include simple geometrical changes such as translations (shifting pixels), rotations, horizontal flips, and zooming in or out. The goal is to modify the images to prevent a CNN model from overfitting by exposing it to a broader range of variations, thereby increasing the variety and diversity of the training set.

The **ImageDataGenerator** class generates batches of tensor image data with real-time data augmentation.

```
from google.colab import drive
drive.mount('/content/drive')
```

```
# Creating an instance of the ImageDataGenerator class for the training set
train_datagen = ImageDataGenerator(rescale = 1./255,         # Rescaling pixel values to be between 0 and 1 (normalization/feature scaling)
                                   shear_range = 0.2,        # Randomly applying shearing transformations
                                   zoom_range = 0.2,         # Randomly zooming into images
                                   horizontal_flip = True)   # Randomly flipping images horizontally


# Generating the training set using the flow_from_directory method
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/Machine Learning A-Z/Part 8 - Deep Learning/2 Convolutional Neural Networks (CNN)/dataset/training_set',
                                                 target_size = (64, 64),  # Resizing all images to 64x64 pixels
                                                 batch_size = 32,         # Number of images to be yielded from the generator per batch
                                                 class_mode = 'binary')   # Setting the class mode for binary classification (e.g., cat vs. dog)
```

### Preprocessing the Test set

Unlike the training set, the test images should not undergo transformations such as shearing, zooming, or flipping because they need to remain as they are, simulating new, unseen images like those encountered in production.

However, the test images should still be rescaled (feature scaling) to ensure consistency with the training set. This is similar to the approach used when applying transformations to training and test sets to avoid information leakage. Keeping the test images intact while scaling ensures the model can correctly predict using the same pixel scaling as the training data.

```
# Creating an instance of the ImageDataGenerator class for the test set
test_datagen = ImageDataGenerator(rescale = 1./255)

# Generating the test set using the flow_from_directory method
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/Machine Learning A-Z/Part 8 - Deep Learning/2 Convolutional Neural Networks (CNN)/dataset/test_set',
                                            target_size = (64, 64),  # Resizing all images to 64x64 pixels
                                            batch_size = 32,
                                            class_mode = 'binary')
```

## Part 2 - Building the CNN

### Initialising the CNN

```
cnn = tf.keras.models.Sequential()
```

### Step 1 - Convolution

```
# Adding the first convolutional layer
#   Number of filters (feature detectors) to apply
#   Size of each filter (3x3 matrix)
#   Activation function to apply (ReLU for non-linearity)
#   Input shape of the images (64x64 pixels with 3 color channels)
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

### Step 2 - Pooling

```
# Adding the first pooling layer
#   Size of the pooling window (2x2 matrix, recommended for max pooling)
#   Step size to move the pooling window, the window will move 2 pixels at a time
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

### Adding a second convolutional layer

```
# Adding the second convolutional layer
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))

# Adding the second pooling layer
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

### Step 3 - Flattening

```
# Flattening the layers to convert them into a single vector
cnn.add(tf.keras.layers.Flatten())
```

### Step 4 - Full Connection

```
# Adding a fully connected (dense) layer
#   Number of neurons in the dense layer
#   Activation function (ReLU) - recommendation: until we reach final layer, use ReLU
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

### Step 5 - Output Layer

```
# Adding the output layer
#   Single neuron output for binary classification
#   Sigmoid activation function for binary classification output
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

## Part 3 - Training the CNN

### Compiling the CNN

To compile the CNN, we need to connect it to an **optimizer**, **loss function**, and **metrics**.

- The **Adam optimizer** is chosen to perform stochastic gradient descent and update the weights to minimize the loss.
- The **binary cross-entropy loss** function is selected, as it's appropriate for binary classification tasks.
- The **accuracy** metric is also used, as it's the most relevant way to measure the model's performance in this scenario.

```
# Compiling the CNN model
#   Optimizer used to adjust the weights (Adam optimizer is commonly used)
#   Loss function for binary classification tasks
#   Metrics used to evaluate the model performance (accuracy in this case)
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Training the CNN on the Training set and evaluating it on the Test set

Note: we are not only training the CNN on the training set, but also evaluating it at the same time on the test set.

The choice of **epochs** for training the CNN:
- Initially, 10 epochs were used, but the accuracy did not converge.
- Next, 15 epochs were tried, but it still wasn't sufficient.
- Finally, 25 epochs were used, which resulted in good convergence for both the training and test sets. Therefore, **25 epochs** is the chosen value for this training process.

```
# Training the CNN model
#   Training data generated by the ImageDataGenerator
#   Validation data (test set) to evaluate the model's performance during training
#   Number of epochs (full passes through the training data)
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

The final accuracy achieved is 90% on the training set and 80% on the test set, which is considered good. Without image augmentation preprocessing, the training accuracy could have reached 98-99%, indicating overfitting, while the test accuracy would have dropped to about 70%. This highlights the importance of image augmentation in preventing overfitting.

## Part 4 - Making a single prediction

```
import numpy as np
from tensorflow.keras.preprocessing import image

# Loading the image to be tested and resizing it to match the input shape of the CNN
test_image = image.load_img('/content/drive/MyDrive/Colab Notebooks/Machine Learning A-Z/Part 8 - Deep Learning/2 Convolutional Neural Networks (CNN)/dataset/single_prediction/cat_or_dog_1.jpg', target_size = (64, 64))

# Converting the image to a numpy array - format expected by the predict method
test_image = image.img_to_array(test_image)

# Expanding dimensions to match the input shape (adding a batch dimension)
test_image = np.expand_dims(test_image, axis = 0)

# Predicting the class of the image using the trained CNN model
result = cnn.predict(test_image)

# Retrieving the class indices from the training set to map the result (e.g., {'cat': 0, 'dog': 1})
training_set.class_indices

if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'
```

**Batch dimension:**

The CNN was trained using **batches of images** rather than single images, which introduces an extra dimension in the data corresponding to the batch size. For instance, each batch can contain 32 images. Even when deploying the model for a **single image** prediction, the image must still be formatted as a batch (even if it only contains one image) to ensure that the CNN recognizes the batch structure. Thus, an extra dimension is added to the input to represent the batch, allowing the CNN's prediction method to function correctly.

**How to determine the class corresponding to the predictions (cats vs. dogs)**

The `class_indices` attribute from the training set object identifies which index corresponds to which classspecifically, **0 for cat** and **1 for dog**. By printing it, we get the answer.

**How to access the predicted result:**

To make a single prediction, the predicted result needs to be accessed from the batch dimension, as predictions are made in batches even if its just one image. The process includes:
1. Accessing the batch (the prediction will be at index **0** since it's the only image).
2. Accessing the single prediction within that batch, which is also at index **0**.
3. Checking the prediction value: if it equals **1**, it indicates a dog; if it equals **0**, it indicates a cat.

```
print(prediction)
```
